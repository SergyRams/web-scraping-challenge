```python
def url_to_jpg(i, url, file_path):
#----------------------------------------------------------------
# PROCEDURE THAT RETRIEVES TO URL and SAVE THE IMAGE AS .jpg File to Defined Directory
    #Args
    
    filename = 'image-{}.jpg'.format(i)
    full_path = '{}{}'.format(file_path, filename)
    urllib.request.urlretrieve(url, full_path)
    
    print('{} saved.'.format(filename))
    
    return None
#------------------------------------------------------------------
# DEFINITION OF GLOBAL VARIABLES READ DATA AND ITERATE THROUGH URLs
#Setting Constants
FILENAME = 'imgs_urls.csv'
FILE_PATH = 'images/'
#Read List of URLs as Pandas DataFrame
urls = pd.read_csv(FILENAME)

#Save Images to the Directory by iterating through the List 
for i, url in enumerate(urls.values):
    url_to_jpg(i, url[0], FILE_PATH)
```

```python
# If you wanted the text only do match=soup.title.text.
#You can pass as an argument class_=  and you use the underscore because class is a reserved word in python

match=soup.find_all('div',class_="rollover_description_inner")
print(match)

#match=soup.find('div',class_='nav_container')
#print(match)

# Retrieve the parent divs for all articles
results = soup.find_all('div', class_="rollover_description_inner")

# loop over results to get article data
for result in results:
    # scrape the article header 
    header = result.find('h1', class_='article-item__headline')
    
    # scrape the article subheader
    subheader = result.find_all('h2', class_='module_title') #.text
    
    # scrape the datetime
    span = result.find_all('span', class_='image_and_description_container')
    
    # get only the date from the datetime
   # date = datetime.split('T')[0]
    
    # print article data
    print('-----------------')
    print(header)
    print(subheader)
    print(span)

    # Dictionary to be inserted into MongoDB
    post = {
        'header': header,
        'subheader': subheader,
        'span': span,
    }

    # Insert dictionary into MongoDB as a document
    collection.insert_one(post)
    
#article =soup.find('div',class_='article')
#print(article)
```

```python
# Retrieve the parent divs for all articles
results = soup.find_all('div', class_='rollover_description_inner')

# loop over results to get article data
for result in results:
    # scrape the article header 
    header = result.find('div', class_='image_and_description_container').text
    
    # scrape the article subheader
    subheader = result.find('h2', class_='article-item__subheader').text
    
    # scrape the datetime
    datetime = result.find('span', class_='article-item__date')['data-date'] 
    
    # get only the date from the datetime
    date = datetime.split('T')[0]
    
    # print article data
    print('-----------------')
    print(header)
    print(subheader)
    print(date)

    # Dictionary to be inserted into MongoDB
    post = {
        'header': header,
        'subheader': subheader,
        'date': date,
    }

    # Insert dictionary into MongoDB as a document
    collection.insert_one(post)
```
